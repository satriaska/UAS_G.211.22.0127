```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
df = pd.read_csv('curah_hujan.csv', sep=';')

# Mengambil semua data
df_subset = df

# Plotting
plt.figure(figsize=(10, 6))

# Mengatur lebar bar
bar_width = 0.15

# Mengatur posisi bar untuk setiap periode curah hujan
bulan = np.arange(len(df_subset))
posisi_1981_1990 = bulan
posisi_1991_2000 = bulan + bar_width
posisi_2001_2010 = bulan + 2 * bar_width

# Bar plot untuk curah hujan tiap periode
plt.bar(posisi_1981_1990, df_subset['Curah Hujan 1981-1990'], color='blue', width=bar_width, label='1981-1990')
plt.bar(posisi_1991_2000, df_subset['Curah Hujan 1991-2000'], color='red', width=bar_width, label='1991-2000')
plt.bar(posisi_2001_2010, df_subset['Curah Hujan 2001-2010'], color='green', width=bar_width, label='2001-2010')

# Konfigurasi plot
plt.xlabel('Bulan')
plt.ylabel('Curah Hujan (mm)')
plt.ylim(0, 700)
plt.title('Curah Hujan per Bulan')
plt.xticks(bulan + bar_width, df_subset['Bulan'])

# Menampilkan legenda menyamping di tengah
plt.legend(bbox_to_anchor=(0.5, 1.15), loc='upper center', ncol=3)

# Tampilkan plot
plt.show()
```
